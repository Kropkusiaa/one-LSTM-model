```python
import kagglehub
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
os.environ['KAGGLEHUB_CACHE'] = '/'
```

```python
path = kagglehub.dataset_download("nphantawee/pump-sensor-data")

print("Path to dataset files:", path)
```

```python
data = pd.read_csv("/datasets/nphantawee/pump-sensor-data/versions/1/sensor.csv")
```

```python
data.head(10)
```

```python
data.isna().sum()
```

```python
data.rename(columns={"Unnamed: 0": "operation_no"}, inplace=True)
```

```python
data.set_index('operation_no', inplace=True)
```

```python
data.head(10)
```

```python
data['time'] = pd.to_datetime(data.pop('timestamp')).dt.time
```

```python
data.head(10)
```

```python
data.insert(0, 'time', data.pop('time'))
data.head(30)
```

```python
data.drop(['sensor_00', 'sensor_15', 'sensor_50', 'sensor_51'], axis=1, inplace=True)
data.head(10)
```

```python
data.isna().sum()
```

```python
data['machine_status'].value_counts()
```

```python
sensors = [c for c in data.columns if c.startswith("sensor_")]
data.loc[data[sensors].isna().any(axis=1), "machine_status"].value_counts()
```

```python
data[sensors].dtypes
```

```python
data.loc[data["machine_status"] == "RECOVERING", sensors] = data.loc[data["machine_status"] == "RECOVERING", sensors].fillna(0)
```

```python
data.loc[data[sensors].isna().any(axis=1), "machine_status"].value_counts()
```

```python
data.loc[data["machine_status"] == "NORMAL", sensors] = (
    data.loc[data["machine_status"] == "NORMAL", sensors]
    .apply(lambda x: x.fillna(x.mean()), axis=0)
)

data.loc[data[sensors].isna().any(axis=1), "machine_status"].value_counts()
```

```python
data.isna().sum()
```

```python
data.loc[data["machine_status"] == "RECOVERING", sensors].describe()
```

```python
data.to_csv("sensors_clean.csv", index=True)
```
